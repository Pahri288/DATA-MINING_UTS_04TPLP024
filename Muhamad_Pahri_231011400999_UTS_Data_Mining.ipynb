```python
hasilgabung =pd.concat([dataset,dataset2,dataset3],ignore_index=True)
```

```python
x = hasilgabung.iloc[:,7: 76]
```

```python
y = hasilgabung.iloc[:,83: 84]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
alya = DecisionTreeClassifier(criterion='entropy', splitter = 'random')
alya.fit(x_train,y_train)
```

```python
y_pred = alya.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize = (10, 7))
tree.plot_tree(alya, feature_names = x.columns.values, class_names = np.array([ 'Benign Traffic','DDos ICMP Flood','DDoS UDP Flood']), filled = True)
plt.show()
```

```python
import pandas as pd

# Asumsikan bahwa 'hasilgabung' merupakan DataFrame hasil penggabungan beberapa dataset secara vertikal
hasilgabung = pd.concat([dataset, dataset2, dataset3], ignore_index=True)

# Ambil jumlah kolom dari DataFrame
jumlah_kolom = hasilgabung.shape[1]

# Tentukan posisi kolom tengah
tengah = jumlah_kolom // 2

# Pisahkan DataFrame menjadi dua bagian berdasarkan kolom
kiri = hasilgabung.iloc[:, :tengah]   # Kolom kiri
kanan = hasilgabung.iloc[:, tengah:]  # Kolom kanan
```

```python
# Tampilkan DataFrame hasilgabung dengan secara keseluruhan
print("DataFrame Gabungan:")
print(hasilgabung)
print("\n")

# Tampilkan bagian kiri dari DataFrame
print("Bagian Kiri:")
print(kiri)
print("\n")

# Tampilkan bagian kanan dari DataFrame
print("Bagian Kanan:")
print(kanan)
```
